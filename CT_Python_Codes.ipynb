```
import random
from PIL import Image

def logic(flowchart, statements, ans, given):
    flowchart = flowchart
    flowchart.show()

    statements = statements
    for statement in statements:
        print("Statement " + str(statement) + ". " + statements[statement])

    ans = ans
    check = {}
    for item in ans:
        check[item] = 0

    display = []
    given = given
    while len(display) < given:
        temp = random.randint(1, 7)
        if temp not in display:
            display.append(temp)

    print("\n")
    for item in ans:
        if ans[item] in display:
            print(item + ": " + str(ans[item]) + ". " + str(statements[ans[item]]))
            check[item] = ans[item]
        else:
            print(item + ": ")

    print("\n")
    for item in ans:
        if ans[item] not in display:
            check[item] = int(input("Enter the statement number matching to " + item + ": "))

    if check == ans:
        print("\nCorrect answer")
    else:
        print("\nWrong answer")
        print("Correct answer is " + str(ans))

def question_55():
    print("Following flowchart represents the stages of a person from the alarm ringing to leaving the bed. Fill in the blacks using the given options.")
    flowchart = Image.open("./Flowcharts/Blank/Alarm Clock.png")
    statements = { 1 : "Snooze", 2 : "5 Minutes", 3 : "Ready to wake up", 4 : "Alarm rings", 5 : "Turn alarm Off", 6 : "Leave bed", 7 : "Will you be late" }
    ans = { "A": 4, "B": 3, "C": 7, "D": 1, "E": 6, "F": 2, "G": 5 }
    given = 2
    logic(flowchart, statements, ans, given)

def question_56():
    print("Following flowchart represents the stages of a person’s decision making process of whether to carry an umbrella while going out. Fill in the blacks using the given options.")
    flowchart = Image.open("./Flowcharts/Blank/Should I Bring an Umbrella.png")
    statements = { 1 : "Found umbrella", 2 : "Don’t carry umbrella", 3 : "Find umbrella", 4 : "Carry umbrella", 5 : "Raining" }
    ans = { "A": 5, "B": 3, "C": 1, "D": 4, "E": 2 }
    given = 2
    logic(flowchart, statements, ans, given)

def question_57():
    print("Following flowchart is designed to advise a person on what will be the best transportation option to travel from city A to city B. Fill in the blacks using the given options.")
    flowchart = Image.open("./Flowcharts/Blank/Travel Plan.png")
    statements = { 1 : "Flight price", 2 : "AC train", 3 : "Arrive at city B", 4 : "Travel from city A to city B", 5 : "Daytime flight", 6 : "Non-AC train", 7 : "Train price ", 8 : "Distance", 9 : "Red eye flight"  }
    ans = { "A": 4, "B": 8, "C": 1, "D": 5, "E": 9, "F": 7, "G": 2, "H": 6, "I": 3 }
    given = 3
    logic(flowchart, statements, ans, given)

def question_58():
    print("Following flowchart represents a sequence of steps through which a patient goes after entering any clinic. Fill in the blacks using the given options.")
    flowchart = Image.open("./Flowcharts/Blank/Medical Services.png")
    statements = { 1 : "Register the patient", 2 : "Need follow up", 3 : "Wait 15 minutes for available nurse", 4 : "Assign patient to doctor and checked-up", 5 : "Need medication", 6 : "Patient arrives", 7 : "Nurse available", 8 : "Wait 15 minutes for available doctor", 9 : "Arrange appointment", 10 : "Registered patient", 11 : "Patient leaves", 12 : "Doctor available", 13 : "Give patient prescription", 14 : "Record health condition" }
    ans = { "A": 6, "B": 10, "C": 1, "D": 7, "E": 3, "F": 14, "G": 12, "H": 8, "I": 4, "J": 2, "K": 9, "L": 5, "M": 13, "N": 11 }
    given = 5
    logic(flowchart, statements, ans, given)

def question_59():
    print("Following flowchart represents how a vacuum robot is programmed to clean a room or a house and take different actions corresponding to different situations. Fill in the blacks using the given options.")
    flowchart = Image.open("./Flowcharts/Blank/Vacuum Robot.png")
    statements = { 1 : "Route", 2 : "Follow route", 3 : "Return to power dock", 4 : "Scan environment", 5 : "Battery low", 6 : "Stop moving", 7 : "Plan route", 8 : "Vacuum Off (Battery low)", 9 : "Power Off", 10 : "Power On", 11 : "Finished route", 12 : "Vacuum Off (Vacuum full)", 13 : "Generate map and location", 14 : "Error indicator On", 15 : "Vacuum full", 16 : "Vacuum On" }
    ans = { "A": 10, "B": 4, "C": 13, "D": 7, "E": 1, "F": 16, "G": 2, "H": 11, "I": 5, "J": 15, "K": 8, "L": 3, "M": 12, "N": 6, "O": 14, "P": 9 }
    given = 5
    logic(flowchart, statements, ans, given)

def question_60():
    print("Following flowchart represents the operation of a vending machine which is dispatching cans. Fill in the blacks using the given options.")
    flowchart = Image.open("./Flowcharts/Blank/Vending Machine.png")
    statements = { 1 : "Get price", 2 : "inserted >= price", 3 : "Get can ID", 4 : "Return change", 5 : "Get coinValue", 6 : "inserted = 0", 7 : "drop can", 8 : "Cans database", 9 : "inserted = inserted + coinValue", 10 : "Receive coin(s)"  }
    ans = { "A": 3, "B": 8, "C": 1, "D": 6, "E": 10, "F": 5, "G": 9, "H": 2, "I": 7, "J": 4 }
    given = 3
    logic(flowchart, statements, ans, given)
    
question_55()
question_56()
question_57()
question_58()
question_59()
question_60()
```
